```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_sim = 100
v_fin = np.zeros([n_sim, 3])
b = np.linspace(0,3,n_sim)
for i in range(n_sim):
    pos = np.load("./XocSimetric"+str(b[i])+"_Sig_1_Eps_1_Niter_7500_tstep_0.001_Integr_Euler/positions.npy")
    vel = np.load("./XocSimetric"+str(b[i])+"_Sig_1_Eps_1_Niter_7500_tstep_0.001_Integr_Euler/velocities.npy")

    v_fin[i] = vel[0,:,-1]
```

```python
x_eix = np.array([1,0,0])
y_eix = np.array([0,1,0])
scalar = np.dot(v_fin,x_eix)
cosangle = np.divide(scalar, np.linalg.norm(v_fin, axis = 1))
sinangle = np.divide(np.dot(v_fin,y_eix), np.linalg.norm(v_fin, axis = 1))
```

```python
plt.plot(b,np.arctan2(sinangle,cosangle))
#plt.plot(np.arccos(cosangle))
#plt.plot(np.arcsin(sinangle))
plt.show()
```
