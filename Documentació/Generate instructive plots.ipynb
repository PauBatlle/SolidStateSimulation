```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1.5,200)
cosy = np.minimum(2*x**2 - 1, 1)
y = np.arccos(cosy)/(2*np.pi)*360
```

```python
plt.plot(2*x,y)
plt.title("Elastic collision of rigid equal spheres")
plt.xlabel("Impact Parameter in proportion to Radius $b/R$")
plt.ylabel("Scattering Angle $ \Theta $ (ยบ)")
plt.yticks(np.arange(y.min(), y.max()+1, 20))
plt.savefig("Billar_collision.jpg")
plt.show()
```

```python
v = np.linspace(0,10, 100)
x = (1+np.sqrt(1+v**2))/2
r = np.power(x, -1/6)
```

```python
plt.plot(v,r)
plt.title("Frontal collision using Lennard-Jones potential")
plt.ylabel("Minimum atom to atom distance $(r_{min}/\sigma$)")
plt.xlabel("Velocity at infinity $v_{\infty}$ (reduced units)")
plt.savefig("Lennard-Jonesradius.jpg")
plt.show()
```
