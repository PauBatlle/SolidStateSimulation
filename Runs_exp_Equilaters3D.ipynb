```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
    
for i in range(100):
    time = 100*i


    xs = posicions[:,0,time]
    ys = posicions[:,1,time]
    zs = posicions[:,2,time]
    ax.scatter(xs, ys, zs, c='b', marker='o', s = 0.1)

    ax.set_aspect('equal')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')


plt.title("Trajectòries de les partícules. Cristall HCP a Temperatura 0.1")
plt.savefig("CristallHCP3D.jpg")
plt.show()
```
